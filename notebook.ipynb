```python
# Temperaturverteilung eindimensionaler Stab

# Parameter
L = 1.0    # Stablänge
k = 1.0    # Wärmeleitfähigkeit
q = 1.0    # Wärmequelle
q_L = 0.0  # Wärmefluss am rechten Ende
T0 = 100.0 # Fixierte Temperatur am linken Rand
n = 3      # Anzahl der Knoten

# Elementgröße
h = L / (n-1)

# Steifigkeitsmatrix K
K = np.array([[ k/h, -k/h, 0 ],
              [-k/h,  2*k/h, -k/h ],
              [ 0, -k/h, k/h ]])

# Lastvektor F
F = np.array([0, q*h, q*h + q_L])

# Randbedingungen einbauen
K[0, :] = 0
K[0, 0] = 1
F[0] = T0

# Gleichungssystem lösen
T = np.linalg.solve(K, F)

print("Temperaturen an den Knotenpunkten:", T)
```
